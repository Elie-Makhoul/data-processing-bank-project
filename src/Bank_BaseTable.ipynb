## 2. Feature Engineering

#### 2.1 Initialize Base Table

##### 2.1.1 create base table

```python
# create copy
client_processed = client.copy()

# create owner table
owner = disp[disp["type"] == "OWNER"]
owner
# create base table 
base_table = pd.merge(client_processed, owner, how="inner", on="client_id")
base_table
```

##### 2.1.2 Adjust Client Birth Number

```python
# Birth year
base_table["birth_year"] = 1900 + base_table["birth_number"].astype(str).str[:2].astype(int)

# birth month
base_table['birth_month'] = base_table['birth_number'].astype(str).str[2:4].astype(int)

# birth day
base_table["birth_day"] = base_table['birth_number'].astype(str).str[-2:].astype(int)

# gender
base_table['gender'] = "M"
base_table.loc[base_table['birth_month'] > 50, "gender"] = "F"

# fix birth month 
base_table.loc[base_table['birth_month'] > 50, 'birth_month'] = base_table.loc[base_table['birth_month'] > 50, 'birth_month'] - 50

base_table.drop("birth_number", axis=1, inplace= True)
```

##### 2.1.3 Compute Age

```python
base_table['age'] = 1996 - base_table['birth_year'] 
```

##### 2.1.4 Compute Age Group

```python
base_table['age_group'] = base_table['age'] // 10 * 10
```

##### 2.1.5 Compute Age Category

```python
base_table['age_group'] = ""
base_table.loc[base_table['age'] <= 14, "age_group"] = "children"
base_table.loc[((base_table['age'] >= 15) & (base_table['age'] <= 24)), "age_group"] = "youth"
base_table.loc[((base_table['age'] >= 25) & (base_table['age'] <= 64)), "age_group"] = "Adult"
base_table.loc[(base_table['age'] >= 65, "age_group")] = "Senior"
```

#### 2.2 Accounts 

##### 2.2.1 accounts created before 1996

```python
# take accounts created before 1996 to have sufficient data
accountbef96 = account.loc[account['date'] < 960101]

# convert to acount date time
accountbef96['account_date'] = pd.to_datetime(accountbef96['date'], format="%y%m%d")

accountbef96['LOR'] = 1996 - (1900 + accountbef96['date'] // 10000) 

# drop date column
accountbef96.drop('date', axis=1, inplace=True)
```

##### 2.2.2 Join account 96 to base table

```python
# join base table and account before 1996
base_table = pd.merge(base_table, accountbef96, how="inner", on= "account_id")

# rename columns
base_table.rename(columns={"district_id_x" : "client_district_id", "district_id_y": "bank_district_id" }, inplace=True)
base_table
```

#### 2.3 Transaction



##### 2.3.1 create transactions for 1996

```python
# create transactions from jan 1 to december 31 1996
trans96= trans.loc[(trans["date"] >= 960101) & (trans['date'] <=961231)]
trans96.dtypes
```

##### 2.3.2 Compute Recency

```python
# create new transaction date column
trans96['trans_date'] = pd.to_datetime(trans96['date'], format="%y%m%d")

# find max date 
max_date = trans96.groupby("account_id")['trans_date'].max().reset_index().rename(columns={"trans_date" : "max_trans_date"})

# compute recency
max_date['recency'] = pd.to_datetime("1996-12-31") - max_date['max_trans_date']

# adjust recency days 
max_date['recency'] = max_date['recency'].dt.days

# merge to base table 
base_table = pd.merge(base_table, max_date, how="left", on= "account_id")
```

##### 2.3.3 Compute Frequency

```python
# calculate frequnecy
num_of_transactions = trans96.groupby("account_id").size().reset_index(name="num_of_transactions")

# merge to base tables 
base_table = pd.merge(base_table, num_of_transactions, how="left", on="account_id")
```

##### 2.3.4 Compute Credit

```python
# compute credit 
total_credit = trans96.loc[trans96['type'] == "PRIJEM"].groupby("account_id")['amount'].sum().reset_index(name="total_credit")

# merge to base table
base_table= pd.merge(base_table,total_credit, on="account_id", how="left")
```

##### 2.3.5 Compute Debit

```python
# include the type VYDAJ and VYBER
total_withdrawal = trans96.loc[trans96['type'].isin(["VYDAJ","VYBER"])].groupby("account_id")['amount'].sum().reset_index(name="total_withdrawal")

# merge to base table
base_table= pd.merge(base_table,total_withdrawal, on="account_id", how="left")
```

##### 2.3.6 Compute net saving rate 

```python
# find saving rate
base_table['net_saving_rate'] = (base_table['total_credit'] - base_table['total_withdrawal']) / base_table['total_credit']

# adjust missing value
base_table['net_saving_rate']  = base_table['net_saving_rate'].fillna(0)
```

##### 2.3.7 Compute Min, Max, Mean of balance

```python
# min balance
min_balance = trans96.groupby("account_id")['balance'].min().reset_index(name="min_balance")

# max balance
max_balance = trans96.groupby("account_id")['balance'].max().reset_index(name="max_balance")

# avg balance
avg_balance = trans96.groupby("account_id")['balance'].mean().reset_index(name="avg_balance")

# Merge
base_table= pd.merge(base_table,min_balance, on="account_id", how="left")
base_table= pd.merge(base_table,max_balance, on="account_id", how="left")
base_table= pd.merge(base_table,avg_balance, on="account_id", how="left")

```

##### 2.3.8 Compute Balance Volatility

```python
# compute balance volatility
balance_volatility = trans96.groupby('account_id')['balance'].std().reset_index(name='balance_volatility')

# merge to base table
base_table = pd.merge(base_table, balance_volatility, on="account_id", how="left")
```

##### 2.3.9 Compute Last Balance

```python
# merge transaction and max date to compute last balance
max_trans_96 = pd.merge(trans96, max_date, on="account_id", how="inner")
max_trans_96

# calculate last balance by setting the transaction date equal to the max date / extracting the last balance
last_balance = max_trans_96.loc[max_trans_96['trans_date'] == max_trans_96['max_trans_date']].groupby("account_id")['balance'].last().reset_index(name="last_balance")

# merge to base table
base_table = pd.merge(base_table, last_balance, on="account_id", how="left")
```

##### 2.3.10 Compute days with negative balance

```python
# sort dates and keep last date
trans96_sorted = trans96[["account_id", "date", "balance"]].sort_values(["account_id", "date"]).drop_duplicates(subset=["account_id", "date"], keep="last")

# create boolean column  to identify whether balance is negative on a given day
trans96_sorted["is_negative"] = trans96_sorted["balance"] < 0

# compute negative balance days for each account
neg_balance_days= trans96_sorted.groupby("account_id")['is_negative'].sum().reset_index(name= "neg_balance_days")


# merge to base table
base_table = pd.merge(base_table, neg_balance_days, on="account_id", how="left")
```

##### 2.3.11 Compute Credit to Debit Ratio Quarterly

Create Quarterly

```python
# Q1
q1 = trans96.loc[(trans96["trans_date"] >= "1996-01-01") & (trans96['trans_date'] <= "1996-03-31")]

# Q2
q2 = trans96.loc[(trans96["trans_date"] >= "1996-04-01") & (trans96['trans_date'] <= "1996-06-30")]
# Q3
q3 = trans96.loc[(trans96["trans_date"] >= "1996-07-01") & (trans96['trans_date'] <= "1996-09-30")]
# Q4
q4 = trans96.loc[(trans96["trans_date"] >= "1996-10-01") & (trans96['trans_date'] <= "1996-12-31")]
```

Create Q1 and Q2 total credit and withdrawal

```python
# for Q1
# credit
q1_total_credit = q1.loc[q1['type']== "PRIJEM"].groupby('account_id')['amount'].sum().reset_index(name='q1_total_credit')

# withdrawal
q1_total_withdrawal = q1.loc[q1['type'].isin(["VYDAJ","VYBER"])].groupby('account_id')['amount'].sum().reset_index(name='q1_total_withdrawal')

# for Q2
# credit
q2_total_credit = q2.loc[q2['type']== "PRIJEM"].groupby('account_id')['amount'].sum().reset_index(name='q2_total_credit')

# withdrawal
q2_total_withdrawal = q2.loc[q2['type'].isin(["VYDAJ","VYBER"])].groupby('account_id')['amount'].sum().reset_index(name='q2_total_withdrawal')

# for Q3
# credit
q3_total_credit = q3.loc[q3['type']== "PRIJEM"].groupby('account_id')['amount'].sum().reset_index(name='q3_total_credit')

# withdrawal
q3_total_withdrawal = q3.loc[q3['type'].isin(["VYDAJ","VYBER"])].groupby('account_id')['amount'].sum().reset_index(name='q3_total_withdrawal')

# for Q4
# credit
q4_total_credit = q4.loc[q4['type']== "PRIJEM"].groupby('account_id')['amount'].sum().reset_index(name='q4_total_credit')

# withdrawal
q4_total_withdrawal = q4.loc[q4['type'].isin(["VYDAJ","VYBER"])].groupby('account_id')['amount'].sum().reset_index(name='q4_total_withdrawal')
```

Merge credit and withdrawal for each quarter

```python
# q1 merge
q1_credit_debit = pd.merge(q1_total_credit, q1_total_withdrawal, on="account_id", how="inner")

# q2 merge
q2_credit_debit = pd.merge(q2_total_credit, q2_total_withdrawal, on="account_id", how="inner")

# q3 merge
q3_credit_debit = pd.merge(q3_total_credit, q3_total_withdrawal, on="account_id", how="inner")

# q4 merge 
q4_credit_debit = pd.merge(q4_total_credit, q4_total_withdrawal, on="account_id", how="inner")
```

compute credit to debit ratio for each quarter

```python
# Calculate Q1 credit to debit_ratio
q1_credit_debit['q1_credit_debit_ratio'] = q1_credit_debit['q1_total_credit'] / q1_credit_debit['q1_total_withdrawal']
q1_credit_debit_ratio= q1_credit_debit[['account_id', 'q1_credit_debit_ratio']]

#   Calculate Q2 credit_debit_ratio
q2_credit_debit['q2_credit_debit_ratio'] = q2_credit_debit['q2_total_credit'] / q2_credit_debit['q2_total_withdrawal']
q2_credit_debit_ratio= q2_credit_debit[['account_id', 'q2_credit_debit_ratio']]

# Calculate Q3 credit_debit_ratio
q3_credit_debit['q3_credit_debit_ratio'] = q3_credit_debit['q3_total_credit'] / q3_credit_debit['q3_total_withdrawal']
q3_credit_debit_ratio= q3_credit_debit[['account_id', 'q3_credit_debit_ratio']]

# Calculate Q4 credit_debit_ratio
q4_credit_debit['q4_credit_debit_ratio'] = q4_credit_debit['q4_total_credit'] / q4_credit_debit['q4_total_withdrawal']
q4_credit_debit_ratio= q4_credit_debit[['account_id', 'q4_credit_debit_ratio']]
```

merge quarterly credit to debit ratio to base table

```python
# merge to q1_credit_debit_ratio to base table
base_table = pd.merge(base_table, q1_credit_debit_ratio, on="account_id", how="left")

# adjust missing values 
base_table['q1_credit_debit_ratio'] = base_table['q1_credit_debit_ratio'].fillna(-1)

# merge to q2_credit_debit_ratio to  base table
base_table = pd.merge(base_table, q2_credit_debit_ratio, on="account_id", how="left")

# adjust missing values
base_table['q2_credit_debit_ratio'] = base_table['q2_credit_debit_ratio'].fillna(-1)


# merge to q3_credit_debit_ratio to  base table
base_table = pd.merge(base_table, q3_credit_debit_ratio, on="account_id", how="left")

# adjust missing values
base_table['q3_credit_debit_ratio'] = base_table['q3_credit_debit_ratio'].fillna(-1)

# merge to q4_credit_debit_ratio to base table
base_table = pd.merge(base_table, q4_credit_debit_ratio, on="account_id", how="left")

# adjust missing values
base_table['q4_credit_debit_ratio'] = base_table['q4_credit_debit_ratio'].fillna(-1)

```

```python
# merge duration to base table
base_table = pd.merge(base_table, loan96[['account_id', "duration"]], on='account_id', how="left")
```

##### 2.5.7 Compute Annual and Monthly  payments

```python
# rename it to monthly payments
loan96.rename(columns={"payments": "monthly_payments"}, inplace=True)

# merge monthly payment to base table
base_table = pd.merge(base_table, loan96[['account_id',  "monthly_payments"]], on='account_id', how="left")

# adjust missing values to zero
base_table['monthly_payments'] = base_table['monthly_payments'].fillna(0)
```

##### 2.5.8 Compute prevoius loan status

```python
# merge monthly payment to base table
base_table = pd.merge(base_table, loan96[['account_id', "status"]], on='account_id', how="left")

# adjust missing values to zero
base_table['status'] = base_table['status'].fillna(-1)

# rename column
base_table.rename(columns={"status":"loan_status"})
```

##### 2.5.9 Compute debt to income ratio

```python
base_table['debt_income_ratio'] = base_table['amount'] / base_table['total_credit']

# adjust missing values to zero
base_table['total_credit'] = base_table['total_credit'].fillna(0)
base_table['debt_income_ratio'] = base_table['debt_income_ratio'].fillna(-1)
```

```python
# compute days from account to loan 
base_table["days_from_acct_to_loan"] = base_table["loan_iss_date"] - base_table['account_date']

# adjust missing values
base_table["days_from_acct_to_loan"] = base_table["days_from_acct_to_loan"].fillna(-1)
```

##### 2.5.11 Compute Avg Balance before and after Loan issued

```python
# to compute avg balance before and after loan issuance i need transactions before 1997

# create transactions before 1997
transbef97= trans.loc[(trans["date"] <= 961231)]

# adjust transaction date
transbef97['trans_date'] = pd.to_datetime(transbef97['date'], format="%y%m%d")

# merge trans to loan table  
transloanbef97 = pd.merge(transbef97, loan96, on="account_id", how='left')

# transaction before loan
trans_bef_loan = transloanbef97.loc[transloanbef97['trans_date'] < transloanbef97['loan_iss_date']]

# compute avg balance for trans_bef_loan
avg_balance_bef_loan = trans_bef_loan.groupby("account_id")['balance'].mean().reset_index(name="avg_balance_bef_loan")

# transaction after loan
trans_aft_loan = transloanbef97.loc[transloanbef97['trans_date'] > transloanbef97['loan_iss_date']]

# compute avg balance for trans_after_loan
avg_balance_aft_loan = trans_aft_loan.groupby("account_id")['balance'].mean().reset_index(name="avg_balance_aft_loan")

# merge to base table
base_table=pd.merge(base_table, avg_balance_bef_loan, on="account_id", how="left")
base_table=pd.merge(base_table, avg_balance_aft_loan, on="account_id", how="left")

# adjust missing values
base_table['avg_balance_bef_loan'] = base_table['avg_balance_bef_loan'].fillna(0)
base_table['avg_balance_aft_loan'] = base_table['avg_balance_aft_loan'].fillna(0)
```

##### 2.5.12 Compute Aggressiveness

```python
base_table['aggressiveness_ratio'] = base_table["amount"] / base_table['duration']

# adjust missing values
base_table['amount'] = base_table['amount'].fillna(0)
base_table['duration'] = base_table['duration'].fillna(0)
base_table['aggressiveness_ratio'] = base_table['aggressiveness_ratio'].fillna(-1)
```
